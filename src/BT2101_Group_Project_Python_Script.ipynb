# Data Preprocessing

## Import Splitted Dataset

```python
df_train = pd.read_csv('cards_train.csv')
df_test = pd.read_csv('cards_test.csv')
```

## Checking For Missing Values

```python
df_train.isnull().any().sum()
df_test.isnull().any().sum()
```

## Data Validation

```python
# remove outliers according to z-scores

from scipy import stats
#df_train = df_train[(z < 3).all(axis=1)]
df_train['LIMIT_BAL'] = df_train['LIMIT_BAL'][(np.abs(stats.zscore(df_train['LIMIT_BAL'])) < 3)]
df_test['LIMIT_BAL'] = df_test['LIMIT_BAL'][(np.abs(stats.zscore(df_test['LIMIT_BAL'])) < 3)]

for i in range(12, 24):
    df_train.iloc[:, i][(np.abs(stats.zscore(df_train.iloc[:, i])) < 3)]
    df_test.iloc[:, i][(np.abs(stats.zscore(df_test.iloc[:, i])) < 3)]

df_train = df_train.dropna(axis = 0)
df_test = df_test.dropna(axis = 0)
```

## Data Transformation

```python
# standardize the coninuous features 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_train['LIMIT_BAL'] = sc.fit_transform(df_train['LIMIT_BAL'].values.reshape(-1,1))
df_test['LIMIT_BAL'] = sc.transform(df_test['LIMIT_BAL'].values.reshape(-1,1))
df_train['AGE'] = sc.fit_transform(df_train['AGE'].values.reshape(-1,1))
df_test['AGE'] = sc.transform(df_test['AGE'].values.reshape(-1,1))

for i in range(7, 24):
    df_train.iloc[:, i] = sc.fit_transform(df_train.iloc[:, i].values.reshape(-1,1))
    df_test.iloc[:, i] = sc.transform(df_test.iloc[:, i].values.reshape(-1,1))
```

## Data Encoding For Discrete Attributes

```python
from sklearn.preprocessing import LabelEncoder
from category_encoders import BinaryEncoder,OneHotEncoder

nom_cat_columns = ["MARRIAGE","SEX"]
ord_cat_columns = ["EDUCATION"]
cat_columns = nom_cat_columns + ord_cat_columns

# get the dataframe without catgorical features
df_train_processed = df_train[[col for col in df_train.columns 
                             if col not in cat_columns]]

# For each categorical column
# We fit a label encoder, transform our column and 
# add it to our new dataframe
label_encoders = {}
for col in cat_columns:
    print("Encoding {}".format(col))
    new_le = LabelEncoder()
    df_train_processed[col] = new_le.fit_transform(df_train[col])
    label_encoders[col] = new_le
```

```python
# use one_hot_encoder to encode nominal features
ohe = OneHotEncoder(handle_unknown='ignore', cols = nom_cat_columns)
df_train_processed = ohe.fit_transform(df_train_processed)

# use binary_encoder to encode ordinal features
be = BinaryEncoder(handle_unknown='ignore', cols = ord_cat_columns)
df_train_processed = be.fit_transform(df_train_processed)

# do the same encoding on test data
df_test_processed = df_test[[col for col in df_test.columns 
                             if col not in cat_columns]]
label_encoders = {}
for col in cat_columns:
    print("Encoding {}".format(col))
    new_le = LabelEncoder()
    df_test_processed[col] = new_le.fit_transform(df_test[col])
    label_encoders[col] = new_le
df_test_processed = ohe.fit_transform(df_test_processed)
df_test_processed = be.fit_transform(df_test_processed)
```

# Feature Selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

x_train = df_train_processed.drop(['ID','default payment next month'],axis = 1).values
y_train = df_train_processed['default payment next month'].values
x_test = df_test_processed.drop(['ID','default payment next month'],axis = 1).values
y_test = df_test_processed['default payment next month'].values

for i in range(1,7):
    selector = SelectKBest(f_classif, k=i*5)
    x_train_processed = selector.fit_transform(x_train, y_train)
    x_test_processed = selector.transform(x_test)

    classifier = RandomForestClassifier(n_jobs=2, random_state=0)

    classifier.fit(x_train_processed, y_train)
    y_pred = classifier.predict(x_test_processed)
    print(str(5*i) + ' : Accuracy ' + str(accuracy_score(y_test, y_pred)))
```

```python
selector = SelectKBest(f_classif, k=10)
x_train_nn = selector.fit_transform(x_train, y_train)
x_test_nn = selector.transform(x_test)

selector = SelectKBest(f_classif, k=5)
x_train = selector.fit_transform(x_train, y_train)
x_test = selector.transform(x_test)
```

## Support Vector Machine

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from time import time
def train_classifier(clf, X_train, y_train):
    clf.fit(X_train, y_train)

def predict_labels(clf, features, target):
    y_pred = clf.predict(features)
    return accuracy_score(target, y_pred, normalize = True)

def train_predict(clf, X_train, y_train, X_test, y_test):
    # Train the classifier
    train_classifier(clf, X_train, y_train)
    print ("accuracy score for test set: {:.5f}.".format(predict_labels(clf, X_test, y_test)))

def evaluate(clf, X_train, y_train, X_test, y_test):
    train_classifier(clf, X_train, y_train)
    y_pred = clf.predict(X_test)
    #accuracy
    print ("accuracy score for test set: {:.5f}.".format(predict_labels(clf, X_test, y_test)))
    print ("accuracy score for training set: {:.5f}.".format(predict_labels(clf, X_train, y_train)))
    # f1 score
    print ("f1 score for training set: {:.5f}.".format(f1_score(y_test, y_pred)))
    # ROC
    print("ROC curve")
    fpr, tpr, threshold = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)
    import matplotlib.pyplot as plt
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.5f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    # False Positive Rate
    print("False Positive Rate:", fpr[1])
```

```python
from sklearn.svm import SVC
# Linear Kernel
print("\nlinearSVM")
svclassifier = SVC(kernel='linear', C = 0.1)
train_predict(svclassifier, x_train, y_train, x_test, y_test)

# Polynomial Kernel
print("\npolynomialkernalSVM")
svclassifier = SVC(kernel='poly', C = 0.1, degree = 4)
train_predict(svclassifier, x_train, y_train, x_test, y_test)

# Gaussian Kernel
print("\ngaussiankernalSVM")
svclassifier = SVC(kernel='rbf', C = 0.2,gamma = 0.1)
train_predict(svclassifier, x_train, y_train, x_test, y_test)

# Sigmoid Kernel
print("\nsigmoidkernalSVM")
svclassifier = SVC(kernel='sigmoid', C = 0.1, gamma = 1, coef0 = 0.1)
train_predict(svclassifier, x_train, y_train, x_test, y_test)
```

```python
# Gaussian Kernel
print("\ngaussiankernalSVM")
svclassifier = SVC(kernel='rbf', C = 0.2,gamma = 0.1)
evaluate(svclassifier, x_train, y_train, x_test, y_test)
```
